Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?

ANS: You should use Stratified Sampling.

Reason:
The population is divided into two distinct groups (strata): boys (60%) and girls (40%). Stratified sampling allows you to select students from each group separately so that both genders are equally represented in the sample.

Simple random sampling might result in more boys than girls (or vice versa), since the population proportions are unequal.

Answer:
Use Stratified Sampling because it ensures equal representation of both boys and girls in the sample.

Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm.Find the sampling error.

ANS: Sampling error is the difference between the sample mean and the population mean.

Given:

Population mean = 160 cm

Sample mean = 158 cm

Sampling Error = Sample Mean - Population Mean

= 158 - 160 = -2 cm

Answer:
The sampling error is -2 cm (the sample mean is 2 cm less than the population mean).
